Simple script for loading in all of the Nursing Home Inspect tables and sending them to the designated MySQL database.

```python
import pandas as pd
import fnmatch
import os
```

```python
#load environment variables
mysql_address = os.environ['MYSQL_ADDRESS']
mysql_user = os.environ['MYSQL_USER']
mysql_password = os.environ['MYSQL_PASSWORD']
mysql_port = os.environ['MYSQL_PORT']
```

```python
#Read in Excel spreadsheets with Pandas
dataframes = []
for f_path in os.listdir('data'):
    if fnmatch.fnmatch(f_path, 'text2567_20171101_cms_reg*u.xlsx'):
        dataframes.append(pd.read_excel('data/' + f_path))
df = pd.concat(dataframes)
```
